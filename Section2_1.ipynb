```python
import itertools
y = list(itertools.permutations([1,1,3,1]))
```

```python
l = list(y)
```

```python
sum = 0
a = []
for i in range(24) :
    print(l[i])
    for j in range(0,4):
        
        if j == 0:
            sum = sum+1
            print(j, sum)
            
        elif l[i][j] <= l[i][j-1]:
            
            print("A")
            sum = sum + 1
            print(j, sum)
            
        elif l[i][j] >= l[i][j-1]:
            
            #print("B",m[i][j] , m[i][j-1])
            p=j
            for k in range(p-1,-1,-1):
                print('B')
                
                if ((l[i][j] > l[i][k]) and (k !=0)):
                    sum = sum + 1
                    print(j, sum)
                    
                if ((l[i][j] > l[i][k]) and (k ==0)):
                    sum = sum + 2
                    print(j, sum)
    a.append(sum)
```

```python
b=[]
b.append(a[0])
for i in range(1,len(l)) :
    b.append(a[i]-a[i-1])
```

```python
sum = 0
for i in range(len(l)) :
    sum = sum + b[i]
mean = sum / 24
print(mean)
```

```python
dev = 0
sumd = 0

dev = [(x - mean) ** 2 for x in b]

for i in range(len(l)):
    sumd = sumd + dev[i]
    
var = sumd / len(b)
devf = var ** 0.5

print(devf)
```
